## 1. Basic model

```python
infectioRateV1 = []
lowestV1 = []
highestV1 = []
infectors_lowestV1 = []
infectors_highestV1 = []

least_f = []
least_r = []
infectors = []
    

print("Loading Infection rate analysis data V1.........")
for f in range(0,50,1) : 
   
    
    with open("../../data/seir/simulated/Ir/V1/lowest/seir_data"+str(f)+".json", 'r') as fp : 
        epidemic_data = json.load(fp)

    df = pd.read_csv('../../data/seir/simulated/Ir/V1/lowest/infection_nb'+str(f)+'.csv')

    lowestV1.append([epidemic_data['S'], epidemic_data['E'],epidemic_data['I'],epidemic_data['R'],
                 epidemic_data['fragile'], epidemic_data['robust']])
    infectors_lowestV1.append([df.values])

    with open("../../data/seir/simulated/Ir/V1/highest/seir_data"+str(f)+".json", 'r') as fp : 
        epidemic_data = json.load(fp)

    df = pd.read_csv('../../data/seir/simulated/Ir/V1/highest/infection_nb'+str(f)+'.csv')


    
    highestV1.append([epidemic_data['S'], epidemic_data['E'],epidemic_data['I'],epidemic_data['R'],
                 epidemic_data['fragile'], epidemic_data['robust']])
    infectors_highestV1.append([df.values])

    least_f.append(epidemic_data['fragile'][-1])
    least_r.append(epidemic_data['robust'][-1])


    
data_lowestV1 = np.array(lowestV1)

means = np.mean(data_lowestV1, axis=0)
print(means.shape, means)
seir_data_lowV1 = {
    'S': means[0], 
    'E': means[1], 
    'I': means[2], 
    'R': means[3],
    'fragile': means[4], 
    'robust': means[5]
}
data_highestV1 = np.array(highestV1)

means = np.mean(data_highestV1, axis=0)
print(means.shape, means)
seir_data_highV1 = {
    'S': means[0], 
    'E': means[1], 
    'I': means[2], 
    'R': means[3],
    'fragile': means[4], 
    'robust': means[5]
}


data_noepi_V1 = []
infectors_noepi_V1 = []
for f in range(0,50,1) : 
   
    
    with open("../../data/seir/simulated/Ir/V1/noepi/seir_data"+str(f)+".json", 'r') as fp : 
        epidemic_data = json.load(fp)
    if epidemic_data['fragile'][-1] <4000 : 
        
        data_noepi_V1.append([epidemic_data['S'], epidemic_data['E'],epidemic_data['I'],epidemic_data['R'],
                 epidemic_data['fragile'], epidemic_data['robust']])
    
    df = pd.read_csv('../../data/seir/simulated/Ir/V1/noepi/infection_nb'+str(f)+'.csv')

    infectors_noepi_V1.append([df.values])  


df_ratio_attackV1 = pd.read_csv("../../data/seir/simulated/ratio_attackV1.csv", header=0, index_col=0)

df_ratio_attackV1    

print("End loading")
```

# 3.1 Load data

```python
low_Rt_r = np.loadtxt("../../data/seir/Rt/lowest/Rt_robust_low.txt")  
low_Rt_f = np.loadtxt("../../data/seir/Rt/lowest/Rt_fragile_low.txt")

        
high_Rt_r =  np.loadtxt("../../data/seir/Rt/highest/Rt_robust_high.txt")
high_Rt_f = np.loadtxt("../../data/seir/Rt/highest/Rt_fragile_high.txt")


with open("../../data/seir/It/lowest/distIT_robust_low.json", "r") as ft : 
    distIT_robust_low = json.load(ft)
with open("../../data/seir/It/highest/distIT_robust_high.json", "r") as ft : 
    distIT_robust_high = json.load(ft)
with open("../../data/seir/It/lowest/distIT_fragile_low.json", "r") as ft : 
    distIT_fragile_low = json.load(ft)
with open("../../data/seir/It/highest/distIT_fragile_high.json", "r") as ft : 
    distIT_fragile_high = json.load(ft)
```
